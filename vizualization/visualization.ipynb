```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
def f(R1=1, R2=2, A=0.1, B=0.1, N=300, M=300):
    # Задаем границы кольца
    # R1, R2 = 1, 3
    
    # Создаем массивы для r и theta
    r = np.linspace(R1, R2, N)
    theta = np.linspace(0, 2 * np.pi, M)
    
    # Создаем сетку полярных координат
    R, Theta = np.meshgrid(r, theta)
    # Пример функции: f(r, theta) = sin(3*theta) * exp(-r)
    # A = 0.1
    # B = 0.1
    
    F = B * (1 + R2 / R * (R**2 + R1**2) / (R1**2 + R2**2) * np.cos(Theta)) + A*R1**3 / (2*(R1**4 + R2**4)) * (R**2 - R2**4/(R**2)) * np.sin(2*Theta)
    # Преобразуем в декартовы координаты
    X = R * np.cos(Theta)
    Y = R * np.sin(Theta)
    
    # Построение графика
    plt.figure(figsize=(6, 6))
    vmin = np.min(F)
    vmax = np.max(F)
    c = plt.pcolormesh(X, Y, F, shading='auto', cmap='plasma', vmin=vmin, vmax=vmax)
    plt.colorbar(c, label='Значение функции')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Распределение температуры в кольце')
    plt.axis('equal')
    plt.savefig('analytical.png')
    plt.show()
```

```python
f()
```

```python
def f_n(filename, R1=1, R2=2):
    
    F = np.loadtxt(filename)
    N = F.shape[0]
    M = F.shape[1]
    
    r = np.linspace(R1, R2, N)
    theta = np.linspace(0, 2 * np.pi, M)
    
    # Создаем сетку полярных координат
    R, Theta = np.meshgrid(r, theta)
    
    # Преобразуем в декартовы координаты
    X = R * np.cos(Theta)
    Y = R * np.sin(Theta)
    
    # Построение графика
    plt.figure(figsize=(6, 6))
    vmin = np.min(F)
    vmax = np.max(F)
    c = plt.pcolormesh(X, Y, np.transpose(F), shading='auto', cmap='hot', vmin=vmin, vmax=vmax)
    plt.colorbar(c, label='Значение функции')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Распределение температуры в кольце')
    plt.axis('equal')
    plt.savefig(filename + '.png')
    plt.show()
```

```python
for i in range(10):
    f_n("numerical_" + str(i) + ".txt")
```

```python
f_n("numerical_10.txt", R1=0.1, R2=0.2)
```
